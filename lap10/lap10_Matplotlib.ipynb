```python
import matplotlib.pyplot as plt

# Dữ liệu
labels = ['Điểm 5', 'Điểm 6', 'Điểm 7', 'Điểm 8', 'Điểm 9', 'Điểm 10', 'Điểm khác']
sizes = [20, 10, 25, 35, 15, 10, 9]
colors = ['red', 'blue', 'yellow', 'green', 'purple', 'pink', 'violet']

# Vẽ biểu đồ
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Đảm bảo rằng biểu đồ tròn
plt.title("Phân bố điểm số môn học DV")
plt.show()
```
