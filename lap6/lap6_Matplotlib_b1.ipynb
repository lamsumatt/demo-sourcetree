Cần ba đối số, giá trị trung bình, độ lệch chuẩn của phân phối chuẩn và số lượng giá trị mong muốn.


Ví dụ 1: Tạo danh sách các mảng là đầu vào bắt buộc để tạo boxplot

```python
import numpy as np
import matplotlib.pyplot as plt

#tạo mảng ngẫu nhiên
np.random.seed(10)
collectn_1 = np.random.normal(100,10,200)
collectn_2 = np.random.normal(80,30,200)
collectn_3 = np.random.normal(90,20,200)
collectn_4 = np.random.normal(70,25,200)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

bp =ax.boxplot([collectn_1,collectn_2,collectn_3,collectn_4])
plt.show()
```

Ví dụ 2: Biểu đồ Box plot cơ bản:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 10
flier_low = np.random.rand(10) * -100

data = np.concatenate((spread, center, flier_high, flier_low))

fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)

plt.show()
```

Ví dụ 3: Sử dụng hàm boxplot() của seaborn. Nhìn vào biểu đồ, sẽ dễ dàng kết luận rằng nhóm ‘C’ có giá trị cao hơn các nhóm khác. Tuy nhiên, ta không thể thấy sự phân bố cơ bản của các dấu chấm trong mỗi nhóm, cũng như số lượng quan sát cho mỗi nhóm.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

a = pd.DataFrame({'group': np.repeat("A", 500), 'value': np.random.normal(10, 5, 500)})
b = pd.DataFrame({'group': np.repeat("B", 500), 'value': np.random.normal(13, 1.2, 500)})
c = pd.DataFrame({'group': np.repeat("B", 500), 'value': np.random.normal(18, 1.2, 500)})
d = pd.DataFrame({'group': np.repeat("C", 20), 'value': np.random.normal(24, 4, 20)})
e = pd.DataFrame({'group': np.repeat("D", 100), 'value': np.random.uniform(10, size=100)})

# Use pd.concat to combine DataFrames
df = pd.concat([a, b, c, d, e])

# Define a color palette
palette = {
    'A': 'red',
    'B': 'blue',
    'C': 'green',
    'D': 'purple'
}

# Create the boxplot with the custom color palette
sns.boxplot(x='group', y='value', data=df, palette=palette)

plt.show()
```
