## Hãy vẽ biểu đồ scatter các điểm như biểu đồ sau:



```python
import matplotlib.pyplot as plt

# Dataset 1
x1 = [89, 43, 36, 36, 95, 10, 87, 66, 34, 38, 20, 26, 34, 90, 33]
y1 = [21, 46, 3, 26, 34, 90, 33, 35, 53, 72, 58, 10, 67, 95, 120]

# Dataset 2
x2 = [26, 29, 48, 20, 56, 2, 48, 20, 56, 36, 66, 72, 20, 56, 2, 47, 40]
y2 = [26, 34, 90, 33, 38, 42, 20, 8, 20, 56, 2, 47, 15, 47, 64, 6, 5]

# Vẽ biểu đồ scatter
plt.scatter(x1, y1, color='blue', label='Dataset 1', marker="s")
plt.scatter(x2, y2, color='red', label='Dataset 2', marker="^")

# Thêm tiêu đề và nhãn cho trục
plt.title('Biểu đồ phân tán của 2 dataset')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')

# Thêm chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()



```

# BÀI TẬP 2:
## Vẽ biểu đồ scatter như biểu đồ sau, có 4 màu với các kiểu khác nhau 
## Cho n1, n2, n3, n4: với random(20)



```python
import matplotlib.pyplot as plt
import numpy as np

# Tạo dữ liệu ngẫu nhiên
n1 = np.random.rand(20)
n2 = np.random.rand(20)
n3 = np.random.rand(20)
n4 = np.random.rand(20)

# Vẽ biểu đồ scatter với 4 màu khác nhau và các kiểu khác nhau
plt.scatter(n1, n2, color='red', marker='o', label='Red circles')
plt.scatter(n2, n3, color='blue', marker='x', label='Blue x')
plt.scatter(n3, n4, color='green', marker='s', label='Green squares')
plt.scatter(n4, n1, color='purple', marker='^', label='Purple triangles')


# Thêm tiêu đề và nhãn cho trục
plt.title('Biểu đồ phân tán với 4 màu và các kiểu khác nhau')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')

# Thêm chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```
